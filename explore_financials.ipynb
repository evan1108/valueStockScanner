#### original value investing bible: Security Analysis (Graham)

## Notes

- look for stocks that are at 2/3 or less of their intrinsic value
- compare P/E to sector P/E, relative P/E to itself, 
- earnings value
- earnings yield - inverse of EPS (%) - return you'd receive if all earnings were paid out in dividend instead of reinvested by the company
- inflation and growth of economy are both historically 3% / yr > earnings should grow 6% a year
- consider trailing, not forward, earnings. rely on audited reports of earnings, not finnicky estimates
- best measure of how much money a company is earning: earnings before interest expense and income tax
- interest expense is a function of how much debt a company has
- EBITDA shows how much cash a company is producing

## Definitions
- cash flow = operating earnings = earnings + non-cash expenses (amortization + depreciation)
- amortization: paying off a debt over time in equal installments. Part of each payment goes toward the loan principal, and part goes toward interest.
- free cash flow = cash flow - capital expenditures needed to keep the lights on
- EBITDA = earnings before interest, taxes, depreciation, and amoritization
